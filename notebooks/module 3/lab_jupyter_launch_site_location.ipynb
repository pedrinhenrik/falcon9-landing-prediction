```python
import numpy as np

def haversine_distance(lat1, lon1, lat2, lon2):
    # Raio mÃ©dio da Terra em km
    R = 6371.0

    # Converter graus para radianos
    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)

    # DiferenÃ§as
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad

    # FÃ³rmula de Haversine
    a = np.sin(dlat/2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    distance_km = R * c
    return distance_km

site_lat, site_lon = 28.573255, -80.646895  # KSC LC-39A
target_lat, target_lon = 28.5618571, -80.577366  # ponto de pouso simulado

dist = haversine_distance(site_lat, site_lon, target_lat, target_lon)
print(f"DistÃ¢ncia: {dist:.2f} km")

import folium

# Criar o mapa centrado no site
map_line = folium.Map(location=[site_lat, site_lon], zoom_start=12)

# Adicionar marcador para o site
folium.Marker([site_lat, site_lon], popup='KSC LC-39A', icon=folium.Icon(color='blue')).add_to(map_line)

# Adicionar marcador para o ponto prÃ³ximo
folium.Marker([target_lat, target_lon], popup='Landing Point', icon=folium.Icon(color='green')).add_to(map_line)

# Adicionar linha conectando os dois
folium.PolyLine(locations=[[site_lat, site_lon], [target_lat, target_lon]], color='red').add_to(map_line)

# Mostrar o mapa
map_line
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
